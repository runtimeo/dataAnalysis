```python
import pandas as pd
import random

data = pd.read_csv('./dataset/company.csv')
print(data)
```

```python
data = data.sample(n=1000,random_state=123,axis=0)
print(data)
```

```python
texts = []
name_texts={}
for index, row in data.iterrows():
    tx = row['经营范围'].replace('\u3000','')
    texts.append(tx)
    name_texts[tx]=row['证券名称']
    
print(name_texts)
```

```python
token = {}
clf = pd.DataFrame(pd.read_csv('./dataset/classification.csv'))
for index, row in clf.iterrows():
    if pd.isnull(row['小类'])==False:
        if row['门类'] not in token:
            token[row['门类']]=[]
        token[row['门类']].append(row['名称'])

print(token)
        
```

```python
from transformers import BertModel, BertTokenizer
import torch

# 加载预训练的BERT模型和分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
model = BertModel.from_pretrained('bert-base-chinese')

# 函数：使用BERT生成文本嵌入
def get_bert_embeddings(name_texts,texts):
    embeddings = {}
    for text in texts:
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
        outputs = model(**inputs)
        # 取[CLS]标记的输出作为句子表示
        cls_embedding = outputs.last_hidden_state[:, 0, :].detach().squeeze().numpy()
        embeddings[name_texts[text]]=cls_embedding
    return embeddings
    
embeddings = get_bert_embeddings(name_texts,texts)


```

```python
emb_dict={}
emb_res={}
temp = []
for key in token.keys():
    string = ''.join(token[key]);
    temp.append(string)
    emb_dict[string]=key

emb_res = get_bert_embeddings(emb_dict,temp)
```

```python
import numpy as np
from sklearn.cluster import KMeans
import numpy as np
from sklearn.preprocessing import StandardScaler

init_centers = []
for key in emb_res.keys():
    init_centers.append(emb_res[key])
init_centers = np.array(init_centers)
scaler = StandardScaler()
init_centers = scaler.fit_transform(init_centers)
test = []
for key in embeddings.keys():
    test.append(embeddings[key])
test = np.array(test)
test = scaler.fit_transform(test)
kmeans = KMeans(n_clusters=20, init=init_centers, n_init=1, random_state=42)
kmeans.fit(test)
labels = kmeans.labels_

final_centers = kmeans.cluster_centers_

print("Cluster labels:", labels)
print("Final cluster centers:\n", final_centers)
```

```python
dist = []
for emb in final_centers:
    distance = np.linalg.norm(emb - init_centers[2])
    dist.append(distance)
dist = np.array(dist)
index = np.argmin(dist)
print(index)
```

```python
samples_in_cluster = [test[i] for i in range(len(test)) if labels[i] == index]
print(samples_in_cluster)
```

```python
lt = []
values=[]
keys = []
scaler = StandardScaler()
samples_in_cluster = np.array(samples_in_cluster)
for key,value in embeddings.items():
    values.append(value)
    keys.append(key)
values = np.array(values)
values = scaler.fit_transform(values)
for idx, value in enumerate(values):
    if np.any(np.all(value == samples_in_cluster, axis=1)):
        lt.append(keys[idx])
print(lt)
```
